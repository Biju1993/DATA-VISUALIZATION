#columns Plot layout



```python
from bokeh.layouts import column
x=np.arange(0,np.pi*2,0.05)
y1=np.sin(x)
y2=np.cos(x)

fig1=figure(plot_width=500,plot_height=500)
fig1.line(x,y1)

fig2=figure(plot_width=500,plot_height=500)
fig2.line(x,y2)

c=column(children=[fig1,fig2],sizing_mode="stretch_both")
output_file('columns plot.html')
show(c)
```

# ROW PLOT LAYOUT FUNCTION

```python
from bokeh.layouts import row
x=np.arange(0,np.pi*2,0.05)
y1=np.sin(x)
y2=np.cos(x)

fig1=figure(plot_width=500,plot_height=500)
fig1.line(x,y1)

fig2=figure(plot_width=500,plot_height=500)
fig2.line(x,y2)

c=row(children=[fig1,fig2],sizing_mode="stretch_both")
output_file('row_plot_layout.html')
show(c)
```

#Grid Plot function

```python
from bokeh.layouts import gridplot
x=np.arange(0,np.pi*2,0.05)
y1=np.sin(x)
y2=np.cos(x)

fig1=figure(plot_width=500,plot_height=500)
fig1.line(x,y1)

fig2=figure(plot_width=500,plot_height=500)
fig2.line(x,y2)

c=gridplot(children=[[fig1,fig2]],sizing_mode="stretch_both")
output_file('grid_plot.html')
show(c)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import matplotlib.animation as animation
#Plotting the tri colours in national flag
# Ref: https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Rectangle.html
green = patch.Rectangle((0,1), width=12, height=2, facecolor='green', edgecolor='black') 
white = patch.Rectangle((0,3), width=12, height=2, facecolor='white', edgecolor='black')
saffron = patch.Rectangle((0,5), width=12, height=2, facecolor='#FF9933', edgecolor='black')
ax,fig = plt.subplots()
fig.add_patch(saffron)
fig.add_patch(white)
fig.add_patch(green)
#PlottingAshokChakra Circle
radius=0.8
plt.plot(6,4, marker = 'o', markerfacecolor = '#000088ff', markersize = 2)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Circle.html?highlight=circle#matplotlib.patches.Circle
chakracircle = patch.Circle((6, 4), radius, color='#000088ff', fill=False, linewidth=4)
fig.add_artist(chakracircle)
#Creating  AshokChakra using Parallogram
# https://byjus.com/maths/parallelogram/
for i in range(0,24):
   x1 = 6 + radius/2 * np.cos(np.pi*i/12 + np.pi/48)
   x2 = 6 + radius/2 * np.cos(np.pi*i/12 - np.pi/48)
   y1 = 4 + radius/2 * np.sin(np.pi*i/12 + np.pi/48)
   y2 = 4 + radius/2 * np.sin(np.pi*i/12 - np.pi/48)
   t = 6 + radius * np.cos(np.pi*i/12)
   u = 4 + radius * np.sin(np.pi*i/12)
   fig.add_patch(patch.Polygon([[6,4], [x1,y1], [t,u],[x2,y2]], fill=True, closed=True, color='#000088ff'))
plt.savefig("National Flag.png",dpi=800)
plt.show()
```
